```python
import tensorflow as tf
from keras.src.legacy.preprocessing.image import ImageDataGenerator
from keras.src.optimizers import Adam
import cv2
import os
import numpy as np
from keras.src.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from keras.src.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from keras.src.applications.vgg16 import VGG16
import matplotlib.pyplot as plt
from keras.src.models import Sequential
from keras.src.layers import Dense, Conv2D , MaxPooling2D , Flatten , Dropout 
from keras.src.models import Model 

# Define labels and image size for both models
labels_type = ['drawings', 'engraving', 'iconography', 'painting', 'sculpture']
labels_style = ['Japanese_Art', 'Neoclassicism', 'Primitivism']
img_size_type = 64  # Image size for the type classifier
img_size_style = 128  # Image size for the style classifier

def get_data(data_dir, labels, img_size):
    """
    Reads images from the specified directory, resizes them,
    and assigns class labels based on the folder structure.
    """
    data = []
    counting_error = 0
    for label in labels:
        path = os.path.join(data_dir, label)
        class_num = labels.index(label)
        for img in os.listdir(path):
            try:
                img_arr = cv2.imread(os.path.join(path, img))[..., ::-1]  # Convert BGR to RGB format
                resized_arr = cv2.resize(img_arr, (img_size, img_size))  # Resize images to preferred size
                data.append((resized_arr, class_num))
            except Exception as e:
                counting_error += 1
    return data

# Load data for the type classifier (5 classes)
train_type = get_data('typedata/training_set/', labels_type, img_size_type)
val_type = get_data('typedata/validation_set/', labels_type, img_size_type)

# Load data for the style classifier (3 classes)
train_style = get_data('styledata/train/', labels_style, img_size_style)
val_style = get_data('styledata/valid/', labels_style, img_size_style)

# Prepare data for the type classifier
x_train_type = np.array([x[0] for x in train_type])
y_train_type = np.array([x[1] for x in train_type])
x_val_type = np.array([x[0] for x in val_type])
y_val_type = np.array([x[1] for x in val_type])

# Normalize the data for the type classifier
x_train_type = x_train_type / 255.0
x_val_type = x_val_type / 255.0

# Data augmentation for the type classifier
datagen_type = ImageDataGenerator(
    rotation_range=30,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

val_datagen_type = ImageDataGenerator()

# Build the type classifier model
model_type = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(img_size_type, img_size_type, 3)),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(len(labels_type), activation='softmax')
])

# Print the summary of the type classifier model
model_type.summary()

# Compile the type classifier model
model_type.compile(optimizer=Adam(learning_rate=0.0001),
                   loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
                   metrics=['accuracy'])

# Callbacks for the type classifier model
callbacks_type = [
    EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=0.00001),
    ModelCheckpoint('best_model_type.keras', monitor='val_accuracy', save_best_only=True)
]

# Train the type classifier model
history_type = model_type.fit(datagen_type.flow(x_train_type, y_train_type, batch_size=32),
                              epochs=50,
                              validation_data=val_datagen_type.flow(x_val_type, y_val_type),
                              callbacks=callbacks_type)

# Plot training & validation accuracy and loss for the type classifier
acc_type = history_type.history['accuracy']
val_acc_type = history_type.history['val_accuracy']
loss_type = history_type.history['loss']
val_loss_type = history_type.history['val_loss']

epochs_range_type = range(len(acc_type))

plt.figure(figsize=(15, 15))
plt.subplot(2, 2, 1)
plt.plot(epochs_range_type, acc_type, label='Training Accuracy')
plt.plot(epochs_range_type, val_acc_type, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy - Type Classifier')

plt.subplot(2, 2, 2)
plt.plot(epochs_range_type, loss_type, label='Training Loss')
plt.plot(epochs_range_type, val_loss_type, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss - Type Classifier')
plt.show()

# Prepare data for the style classifier
x_train_style = np.array([x[0] for x in train_style])
y_train_style = np.array([x[1] for x in train_style])
x_val_style = np.array([x[0] for x in val_style])
y_val_style = np.array([x[1] for x in val_style])

# Normalize the data for the style classifier
x_train_style = x_train_style / 255.0
x_val_style = x_val_style / 255.0

# Data augmentation for the style classifier
datagen_style = ImageDataGenerator(
    rotation_range=30,
    width_shift_range=0.3,
    height_shift_range=0.3,
    shear_range=0.3,
    zoom_range=0.3,
    horizontal_flip=True,
    fill_mode='nearest'
)

val_datagen_style = ImageDataGenerator()

# Build the style classifier model using VGG16 as the base model
base_model_style = VGG16(weights='imagenet', include_top=False, input_shape=(img_size_style, img_size_style, 3))

# Freeze the layers in the base model
for layer in base_model_style.layers:
    layer.trainable = False

# Add custom layers on top of the base model
x = base_model_style.output
x = Flatten()(x)
x = Dense(1024, activation='relu')(x)
x = Dropout(0.5)(x)
x = Dense(1024, activation='relu')(x)
x = Dropout(0.5)(x)
predictions_style = Dense(len(labels_style), activation='softmax')(x)

# Create the final style classifier model
model_style = Model(inputs=base_model_style.input, outputs=predictions_style)

# Print the summary of the style classifier model
model_style.summary()

# Compile the style classifier model
model_style.compile(optimizer=Adam(learning_rate=0.0001),
                    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
                    metrics=['accuracy'])

# Callbacks for the style classifier model
callbacks_style = [
    EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=0.00001),
    ModelCheckpoint('best_model_style.keras', monitor='val_accuracy', save_best_only=True)
]

# Train the style classifier model
history_style = model_style.fit(datagen_style.flow(x_train_style, y_train_style, batch_size=32),
                                epochs=50,
                                validation_data=val_datagen_style.flow(x_val_style, y_val_style),
                                callbacks=callbacks_style)

# Plot training & validation accuracy and loss for the style classifier
acc_style = history_style.history['accuracy']
val_acc_style = history_style.history['val_accuracy']
loss_style = history_style.history['loss']
val_loss_style = history_style.history['val_loss']

epochs_range_style = range(len(acc_style))

plt.figure(figsize=(15, 15))
plt.subplot(2, 2, 1)
plt.plot(epochs_range_style, acc_style, label='Training Accuracy')
plt.plot(epochs_range_style, val_acc_style, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy - Style Classifier')

plt.subplot(2, 2, 2)
plt.plot(epochs_range_style, loss_style, label='Training Loss')
plt.plot(epochs_range_style, val_loss_style, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss - Style Classifier')
plt.show()

# Function to predict the type of artwork
def predict_artwork_type(image_path):
    """
    Predict the type of artwork given an image.

    Args:
    - image_path (str): Path to the image file.

    Returns:
    - predicted_type (str): Predicted type of artwork.
    """
    image = cv2.imread(image_path)[..., ::-1]  # Load image in BGR format, convert to RGB
    resized_image = cv2.resize(image, (img_size_type, img_size_type))  # Resize image
    normalized_image = resized_image / 255.0  # Normalize pixel values

    predictions = model_type.predict(np.expand_dims(normalized_image, axis=0))
    predicted_type = labels_type[np.argmax(predictions[0])]

    return predicted_type

# Function to predict the style of painting
def predict_painting_style(image_path):
    """
    Predict the style of painting given an image.

    Args:
    - image_path (str): Path to the image file.

    Returns:
    - predicted_style (str): Predicted style of painting.
    """
    image = cv2.imread(image_path)[..., ::-1]  # Load image in BGR format, convert to RGB
    resized_image = cv2.resize(image, (img_size_style, img_size_style))  # Resize image
    normalized_image = resized_image / 255.0  # Normalize pixel values

    predictions = model_style.predict(np.expand_dims(normalized_image, axis=0))
    predicted_style = labels_style[np.argmax(predictions[0])]

    return predicted_style

# Example usage
image_path = 'paint-neo.jpg'  # Replace with the path to your image

# Predict the type of the artwork
predicted_type = predict_artwork_type(image_path)
print(f"Predicted Type: {predicted_type}")

# If the predicted type is 'painting', predict the painting style
if predicted_type == 'painting':
    predicted_style = predict_painting_style(image_path)
    print(f"Predicted Painting Style: {predicted_style}")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from keras.api.models import load_model

# Define the labels
labels_type = ['drawings', 'engraving', 'iconography', 'painting', 'sculpture']
labels_style = ['Japanese_Art', 'Neoclassicism', 'Primitivism']

# Define the image sizes
img_size_type = 64
img_size_style = 128

# Load the pre-trained models
model_type = load_model('best_model_type.keras')
model_style = load_model('best_model_style.keras')

# Function to predict the type of artwork
def predict_artwork_type(image_path):
    """
    Predict the type of artwork given an image.

    Args:
    - image_path (str): Path to the image file.

    Returns:
    - predicted_type (str): Predicted type of artwork.
    """
    image = cv2.imread(image_path)[..., ::-1]  # Load image in BGR format, convert to RGB
    resized_image = cv2.resize(image, (img_size_type, img_size_type))  # Resize image
    normalized_image = resized_image / 255.0  # Normalize pixel values

    predictions = model_type.predict(np.expand_dims(normalized_image, axis=0))
    predicted_type = labels_type[np.argmax(predictions[0])]

    return predicted_type

# Function to predict the style of painting
def predict_painting_style(image_path):
    """
    Predict the style of painting given an image.

    Args:
    - image_path (str): Path to the image file.

    Returns:
    - predicted_style (str): Predicted style of painting.
    """
    image = cv2.imread(image_path)[..., ::-1]  # Load image in BGR format, convert to RGB
    resized_image = cv2.resize(image, (img_size_style, img_size_style))  # Resize image
    normalized_image = resized_image / 255.0  # Normalize pixel values

    predictions = model_style.predict(np.expand_dims(normalized_image, axis=0))
    predicted_style = labels_style[np.argmax(predictions[0])]

    return predicted_style

# Function to load an image and display it with the predicted labels
def display_prediction(image_path):
    """
    Display the image and its predicted labels.

    Args:
    - image_path (str): Path to the image file.
    """
    # Predict the type of the artwork
    predicted_type = predict_artwork_type(image_path)
    print(f"Predicted Type: {predicted_type}")

    # Load and display the image
    image = cv2.imread(image_path)[..., ::-1]  # Load image in BGR format, convert to RGB
    plt.figure(figsize=(5, 5))
    plt.imshow(image)
    plt.title(f"Predicted Type: {predicted_type}")
    plt.axis('off')
    plt.show()

    # If the predicted type is 'painting', predict the painting style
    if predicted_type == 'painting':
        predicted_style = predict_painting_style(image_path)
        print(f"Predicted Painting Style: {predicted_style}")

        # Display the image with the predicted style
        plt.figure(figsize=(5, 5))
        plt.imshow(image)
        plt.title(f"Predicted Type: {predicted_type}\nPredicted Style: {predicted_style}")
        plt.axis('off')
        plt.show()

# Example test
image_path = 'Stest.png'  # Replace with the path to your image
display_prediction(image_path)

image_path = 'engravetest.jpg'  # Replace with the path to your image
display_prediction(image_path)

image_path = 'icontest.jpg'  # Replace with the path to your image
display_prediction(image_path)

image_path = 'paint-neo.jpg'  # Replace with the path to your image
display_prediction(image_path)
```
